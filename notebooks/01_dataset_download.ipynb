# EarthScape Dataset Download
Download and stage the Berkeley Earth dataset for backend ingestion.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("berkeleyearth/climate-change-earth-surface-temperature-data")

print("Path to dataset files:", path)
```

```python
from pathlib import Path
import shutil

source_dir = Path(path)
repo_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
target_dir = repo_root / 'data' / 'raw'
target_dir.mkdir(parents=True, exist_ok=True)

copied = 0
for csv_file in source_dir.rglob('*.csv'):
    shutil.copy2(csv_file, target_dir / csv_file.name)
    copied += 1

print(f'Copied {copied} csv files to {target_dir}')
```
